```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def frequency_from_fundamental(n, round_at=0, logs=True):
    
    """Calculate frequency based on a {mass + ruberband} system."""

    frequency = 1/2 * np.sqrt(n**2*T/(m*l))
    
    if round_at:
        frequency = round(frequency, round_at)
        
    if logs:
        print("fundamental frequency : ", 1/2 * np.sqrt(T/(m*l)), "Hz")
        print("frequency", n,  "is", frequency, "Hz")
        
    return frequency
```

```python
m = 0.15 * 10**(-3) #kg 
l = 1.4 * 10**(-3) #m
T = 1 #N

num = 10000
num_harmonics = 3

t = np.linspace(0, 1, num)
frequencies = np.zeros((num_harmonics, num))

for i in range(1, frequencies.shape[0]):
    #create sine wave from frequency
    frequencies[i] = np.sin(frequency_from_fundamental(i) * t)
    plt.plot(t, frequencies[i])

plt.xlim(0, 0.1)
plt.show()
```
