```python
# Importation des librairies r√©unies en une seule cellule, plus pratique
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.io.wavfile import write, read
import wave
import struct
import IPython.display as ipd
```

```python
def visualize_wav(f):

    """Decompose wav file into int16 array. Return a time linspace to plot and the array"""

    framerate, frames = read(f)

    return framerate, frames

framerate, frames = visualize_wav("test.wav")
t_wav = np.arange(0, len(frames)/5)/framerate

plt.figure(figsize = (20,5))

frames1 = frames[ : int(len(frames)/5)]

plt.plot(t_wav, frames1)

plt.show()
```

```python
Framerate, Frames = read('reference.wav')

time1 = np.arange(0, 0.2, 1/Framerate)

plt.figure(figsize = (20,5))

Frames1 = Frames[ : int(0.2*Framerate)]

plt.plot(time1, Frames1)

plt.show()
```
