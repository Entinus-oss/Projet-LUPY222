```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write, read

samplingRate = 44100

def write_melody(writewave):
    """
    écrire la chanson avec chaque son créé par la fonction writewave
    """
    
    fG = 97.999
    fA = 110.000
    fc = 130.813
    fd = 146.832
    fe = 164.814
    fg = 195.998
    fa = 220.000
    fc1 = 261.626


    melody = writewave(fe, 0.6)
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fc1, 0.3))
    melody = np.append(melody, writewave(fc1, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, 0*writewave(fg, 0.6))

    melody = np.append(melody, writewave(fe, 0.6))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fc1, 0.3))
    melody = np.append(melody, writewave(fc1, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, 0*writewave(fg, 0.6))
    
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fa, 0.6))
    melody = np.append(melody, writewave(fa, 0.6))
    melody = np.append(melody, writewave(fg, 1.2))
    
    melody = np.append(melody, writewave(fe, 0.6))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fc, 0.6))
    melody = np.append(melody, writewave(fc, 0.3))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fc, 1.2))

    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fc, 0.3))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fd, 0.9))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.6))
    melody = np.append(melody, writewave(fa, 0.3))
    melody = np.append(melody, writewave(fc1, 0.3))
    melody = np.append(melody, writewave(fg, 1.2))

    melody = np.append(melody, writewave(fd, 0.6))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fg, 0.3))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fc, 0.3))
    melody = np.append(melody, writewave(fA, 0.3))
    melody = np.append(melody, writewave(fG, 1.2))

    melody = np.append(melody, writewave(fA, 0.6))
    melody = np.append(melody, writewave(fc, 0.6))
    melody = np.append(melody, writewave(fd, 0.9))
    melody = np.append(melody, writewave(fe, 0.3))
    melody = np.append(melody, writewave(fc, 0.3))
    melody = np.append(melody, writewave(fd, 0.3))
    melody = np.append(melody, writewave(fc, 0.3))
    melody = np.append(melody, writewave(fA, 0.3))
    melody = np.append(melody, writewave(fG, 1.2))
    melody = np.append(melody, 0*writewave(fe, 1.2))

    wavFile = write("melody2.wav", samplingRate, melody.astype('float32'))
    
    return

def writewave(fr, time):

    amplitudelow = 0.1
    samplingInterval = 1/samplingRate
    
    t = np.arange(0, time, samplingInterval)
    lowFrequencySineWaves = np.zeros((4, t.size))
    frequencies = []
    
    def f(i):
        if i == 0:
            return 2
        if i == 1:
            return 1
        if i == 2:
            return 0
        if i == 3:
            return 4
    
    def add_sine_waves(listOfSineWave, logs=False):
        
        sumOfSineWaves = np.copy(listOfSineWave[0])

        if logs: 
            print("sumOfSineWaves", sumOfSineWaves)
            print(listOfSineWave.ndim)

        for i in range(1, len(listOfSineWave)):
            sumOfSineWaves += listOfSineWave[i]
            #print("sum at", i, ",", sumOfSineWaves)

        if logs:
            print("sumOfSineWaves :", sumOfSineWaves)

        return sumOfSineWaves

    
    for i in range(4):
        #create sine wave from lower frequencies
        frequencies.append((i+1)*fr)
        lowFrequencySineWaves[i] = np.exp(-amplitudelow *i) * np.cos(2*np.pi * (i+1) * fr * t + f(i)*2*np.pi/8)
    
    createdSineWave = add_sine_waves(lowFrequencySineWaves)
    
    createdSineWave *= 1/max(createdSineWave)
    
    return createdSineWave
```
