```python
import wave
import numpy as np
import matplotlib.pyplot as plt
import struct


def writewave1(f, file, framerate, time):
    t = np.arange(0, time, 1.0/framerate)
    wave_data = f(t)
    F = wave.open(file, 'wb')
    F.setnchannels(1)
    F.setsampwidth(2)
    F.setframerate(framerate1)
    for w in wave_data :
        F.writeframes(struct.pack('h', int(w)))
    F.close()
```
